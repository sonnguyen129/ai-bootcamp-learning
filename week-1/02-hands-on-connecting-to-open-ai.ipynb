```python
from openai import OpenAI
from toyaikit.llm import OpenAIClient
from toyaikit.chat import IPythonChatInterface
from toyaikit.chat.runners import OpenAIResponsesRunner

client = OpenAI()

system_prompt = """
You're an assistant that creates bedtime stories. Include emojis in the response.
Ask the user follow up questions and make sure the stories are personalized
""".strip()

llm_client = OpenAIClient(
    model="gpt-4o-mini",
    client=client
)

runner = OpenAIResponsesRunner(
    tools=None,
    developer_prompt=system_prompt,
    chat_interface=IPythonChatInterface(),
    llm_client=llm_client
)
```

```python
runner.run();
```

### Using chat.completions instead of responses

```python
system_prompt = """
you're an assistant that can make jokes. Always find out the name of
the person to make the jokes personalized. Once you know the name,
make the joke about them.
""".strip()

messages = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": "tell me a joke"}
]

response = client.chat.completions.create(
    model='gpt-4o-mini',
    messages=messages
)

response.choices[0].message.content
```

```python
messages.append(response.choices[0].message)
messages.append({"role": "user", "content": "Fredy"})

response = client.chat.completions.create(
    model='gpt-4o-mini',
    messages=messages
)

response.choices[0].message.content
```

When using OpenAI, we should prefer the responses API. It's a newer API and the OpenAI documentation mostly uses it.

However, it's still useful for alternative LLM providers. Most of them follow the OpenAI API, and you can interact with them using the chat.completions API.

For example, this is how you do it with Anthropic:

```python
import os

anthropic_client = OpenAI(
    api_key=os.getenv("ANTHROPIC_API_KEY"),
    base_url="https://api.anthropic.com/v1/"
)

response = anthropic_client.chat.completions.create(
    model='claude-3-5-haiku-latest',
    messages=messages
)

print(response.choices[0].message.content)
```

Try using the respones API:

```python
response = anthropic_client.responses.create(
    model='claude-3-5-haiku-latest',
    input=messages
)
```

```python
import os
from groq import Groq

groq_client = Groq(
    api_key=os.environ.get("GROQ_API_KEY"),
)
```

```python
system_prompt = """
you're an assistant that can make jokes. Always find out the name of
the person to make the jokes personalized. Once you know the name,
make the joke about them.
""".strip()

messages = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": "tell me a joke"}
]
```

```python
response = groq_client.chat.completions.create(
    messages=messages,
    model="llama-3.3-70b-versatile"
)

print(response.choices[0].message.content)
```

```python
messages.append(response.choices[0].message)
messages.append({"role": "user", "content": "Fredy"})
```

```python
response = groq_client.chat.completions.create(
    messages=messages,
    model="llama-3.3-70b-versatile"
)

print(response.choices[0].message.content)
```
